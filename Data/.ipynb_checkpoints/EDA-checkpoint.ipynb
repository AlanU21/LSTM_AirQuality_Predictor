```python
import pandas as pd
```

```python
df = pd.read_excel("cleaned_air_quality_data.xlsx")

df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df.corr()

plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='hour', y='AQI_pm25', data=df)
plt.title('Box Plot of AQI PM2.5 by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('AQI PM2.5')

plt.subplot(1, 2, 2)
sns.boxplot(x='hour', y='AQI_pm10', data=df)
plt.title('Box Plot of AQI PM10 by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('AQI PM10')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
plt.plot(df['Timestamp'], df['pm25_AVERAGE'], label='PM2.5 Average')
plt.title('Time Series of PM2.5 Average')
plt.xlabel('Date')
plt.ylabel('PM2.5 Levels')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(df['Timestamp'], df['pm10'], label='PM10')
plt.title('Time Series of PM10')
plt.xlabel('Date')
plt.ylabel('PM10 Levels')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Timestamp'], df['AQI_pm25'], label='AQI PM2.5')
plt.plot(df['Timestamp'], df['AQI_pm10'], label='AQI PM10', color='red')
plt.title('Time Series Plot for AQI')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.legend()

plt.show()
```
